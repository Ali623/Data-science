```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
df_sur_age = df[df["Survived"] == 1]["Age"]
binz = np.arange(df["Age"].min(),df["Age"].max(),10)
df_sur_age.plot(kind="hist",figsize = (10,5), color = "maroon", bins = binz , rwidth = 0.95 , edgecolor = "black")
plt.style.use("fivethirtyeight")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("hist of survived peoples")
plt.savefig("hist of survived peoples.png")
plt.tight_layout()
plt.show()
```

```python
df_dea_age = df[df["Survived"] == 0]["Age"]
binz = np.arange(df["Age"].min(),df["Age"].max(),10)
df_dea_age.plot(kind="hist",figsize = (10,5), color = "red" , bins = binz , rwidth = 0.95, edgecolor = "black")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("hist of dead peoples")
plt.savefig("hist of dead peoples.png")
plt.tight_layout()
plt.show()
```
